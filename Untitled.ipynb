```python

import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python

img = cv2.imread('res/img1.jpg',0)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
```

```python
original=cv2.imread("./res/img1.jpg")
default=cv2.imread("./res/output.png")

original=cv2.cvtColor(original,cv2.COLOR_BGR2GRAY)
default=cv2.cvtColor(default,cv2.COLOR_BGR2GRAY)

thresh1 = cv2.adaptiveThreshold(original, 255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 199, 5)
thresh2 = cv2.adaptiveThreshold(default, 255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 199, 5)
plt.figure()
plt.imshow(thresh1)
plt.figure()
plt.imshow(thresh2)
```
